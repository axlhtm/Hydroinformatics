## Penggunaan Pandas di Python

Pandas digunakan untuk *data analysis* dan *data manipulation*, dimana fungsi dari Pandas akan menjadi seperti R. Dimana kita dapat mengolah data excel, csv, mqsql yang digunakan dalam cloud computing. 

```python
import pandas as pd
import numpy as np
```

```python
pd.read_csv('kapal_titanic.csv') # Hanya untuk membaca file csv
```

Cara diatas hanya digunakan untuk **membaca** file csv yang kita punya, namun apabila kita ingin **menggunakan atau menganalisa data tersebut, maka kita harus memasukan data tersebut ke dalam suatu variabel**. Seperti contoh dibawah ini.

```python
dataku = pd.read_csv('kapal_titanic.csv')
```

```python
dataku.head()
```

```python
dataku.tail()
```

String diatas adalah contoh untuk mengeksport variable ke dalam suatu file. String 'index=False' mengindikasikan bahwa urutan baris 0-890 tidak dimasukan ke dalam file yang baru.
